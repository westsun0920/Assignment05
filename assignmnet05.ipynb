```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import signal
from skimage import io, color
from skimage import exposure

file_image = 'cau.jpg'

im_color = io.imread(file_image)
im_gray  = color.rgb2gray(im_color)

ker = np.array([[0,0,0],[0,1,0],[0,0,0]]) 
im_conv= signal.convolve2d(im_gray, ker, boundary='symm', mode='same') 

p1 = plt.subplot(2,2,1)
p1.set_title('color image')
plt.imshow(im_color)
plt.axis('off')

p2 = plt.subplot(2,2,2)
p2.set_title('gray image')
plt.imshow(im_gray, cmap='gray')
plt.axis('off')

p3 = plt.subplot(2,2,3)
p3.set_title('convolution kernel')
plt.imshow(ker, cmap='gray')
plt.axis('off')

p4 = plt.subplot(2,2,4)
p4.set_title('convolution result')
plt.imshow(im_conv, cmap='gray')
plt.axis('off')

plt.show()
```

# Derivative in x-direction

```python
kernel_X = np.array([[3,0,-3],[1,0,-1],[2,0,-2]])
image_derX= signal.convolve2d(im_gray, kernel_X, boundary='symm', mode='same') 

p1 = plt.subplot(1,2,1)
plt.imshow(kernel_X, cmap='hsv')
plt.axis('off')

p2 = plt.subplot(1,2,2)
plt.imshow(image_derX, cmap='hsv')
plt.axis('off')
```

# Derivative in y-direction

```python
kernel_Y = np.array([[2,3,2],[-1,-2,-1],[1,-2,1]])
image_derY= signal.convolve2d(im_gray, kernel_Y, boundary='symm', mode='same') 

p1 = plt.subplot(1,2,1)
plt.imshow(kernel_Y, cmap='hsv')
plt.axis('off')
p2 = plt.subplot(1,2,2)
plt.imshow(image_derY, cmap='hsv')
plt.axis('off')
```

# Function for computing the magnitude of the gradient 

```python
def computingMagnitude(vec1,vec2) : 
    return np.sqrt(vec1**2+vec2**2)
```

# Absolute value of gradient

```python
magnitude_grad=computingMagnitude(image_derX,image_derY) 

plt.imshow(magnitude_grad, cmap='hsv')
plt.axis('off')
```

# Function for computing the direction of the gradient 

```python
def computingDirection(vec1,vec2) :
    return np.arctan(vec2/vec1)
```

# Direction of gradient

```python
direction_grad=computingDirection(image_derX,image_derY)

plt.imshow(direction_grad, cmap='hsv')
plt.axis('off')
```

# Smoothing kernel

```python

kernel_smooth = np.array([[1/9,1/9,1/9],[1/9,1/9,1/9],[1/9,1/9,1/9]])
image_smooth= signal.convolve2d(im_gray, kernel_smooth, boundary='symm', mode='same') 

plt.imshow(image_smooth, cmap='gray')
plt.axis('off')
```

# Inverse kernel

```python

kernel_inverse = np.array([[-255,-255,-255],[-255,-255,-255],[-255,-255,-255]])
image_inverse= signal.convolve2d(im_gray, kernel_inverse, boundary='symm', mode='same') 

plt.imshow(image_inverse, cmap='hsv')
plt.axis('off')
```
